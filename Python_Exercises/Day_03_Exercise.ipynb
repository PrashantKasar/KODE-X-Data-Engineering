```python
# DAY 3 PRACTICE – PYTHON
'''
Topics
Loops
Lists
Strings
Dictionaries
Basic Logic Problems

---

Exercise 1 – Sum of Even Numbers

Given a list:

numbers = [3, 8, 12, 5, 7, 10]

Write a program to calculate the sum of only the even numbers in the list.

Expected Output Example:

Sum of even numbers: 30

---
'''
# Exercise 1 – Sum of Even Numbers

numbers = [3, 8, 12, 5, 7, 10]
sum = 0
for i in numbers:
    if i % 2 == 0:
       sum += i
print(f"Sum of Even Number : {sum}")

```

```python
# ================================================================================
'''
Exercise 5 – Character Frequency

Ask the user to enter a word.

Print how many times each character appears.

Example:

Input: apple

Output:

a : 1
p : 2
l : 1
e : 1

Hint: Use dictionary.

# Complete Flow Diagram
User enters word
        │
        ▼
Create empty dictionary {}
        │
        ▼
Read first character
        │
        ▼
Is character already in dictionary?
        │
   ┌────┴────┐
   │         │
  Yes       No
   │         │
Increase    Store
count       count = 1
   │         │
   └────┬────┘
        ▼
Next character?
        │
       Yes
        │
        ▼
Repeat
        │
       No
        ▼
Print every key and its count
---
'''
# Exercise 5 – Character Frequency
word = input("Enter a word: ")
frqe = {}
for i in word:
    if i in frqe:
        frqe[i] += 1
    else:
        frqe[i] = 1
for key, value in frqe.items():
    print(key ,':', value)

```

```python
# ================================================================================
'''
Exercise 7 – Student Marks Dictionary

Create a dictionary:

marks = {
Rahul: 85,
Amit: 72,
Priya: 90,
Neha: 66
}

Print the names of students who scored more than 80.

Expected Output:

Rahul
Priya

---
'''
# Exercise 7 – Student Marks Dictionary
marks = {
    "Rahul": 85,
    "Amit": 72,
    "Priya": 90,
    "Neha": 66
}
for key,value in marks.items():
    if value > 80:
        print(key)

```

```python
# ================================================================================
'''
Exercise 8 – List Length Without len()

Given list:

numbers = [10,20,30,40,50]

Find the length of the list without using len() function.

---
'''
# Exercise 8 – List Length Without len()
numbers = [10,20,30,40,50]

count = 0
for i in numbers:
    count += 1
print(f"Length of list : {count}")

```

```python
# ================================================================================
'''
Exercise 9 – Check Palindrome

Ask user to enter a word.

Check if the word is a palindrome.

Example:

Input: madam
Output: Palindrome

Input: python
Output: Not Palindrome

---
'''
# Exercise 9 – Check Palindrome
word = input("Enter a word: ")

if word == word[::-1]:
    print(f" \'{word}\' word is a palindrome.")
else:
    print(f" \'{word}\' word is not a palindrome.")

```

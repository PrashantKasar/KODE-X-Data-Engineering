```python
# SQL PRACTICE
# Topic: Level 1 + 2 harder multi-concept
#        Focus: NULL, CASE, GROUP BY + HAVING
#        combined with real hospital tables
# SQL Exercise 1 - CASE statement introduction
'''
1. From employees:
   Add a salary_band column using CASE:
   salary >= 85000 → "Senior"
   salary >= 65000 → "Mid"
   salary >= 0     → "Junior"
   NULL salary     → "Unknown"
   Show: name, department, salary, salary_band.

2. From orders:
   Add a status_flag column using CASE:
   "Delivered" → "Completed"
   "Shipped"   → "In Transit"
   "Pending"   → "Waiting"
   "Cancelled" → "Closed"
   Show: order_id, status, amount, status_flag.

3. From appointments:
   Show total fee collected grouped by
   a fee_category derived using CASE:
   fee >= 700 → "High"
   fee >= 400 → "Medium"
   fee < 400  → "Low"
   Use the CASE inside GROUP BY or
   wrap it in a subquery.

---------------------------------------------
'''
# 1
SELECT NAME,
    DEPARTMENT,
    SALARY,
    CASE
        WHEN SALARY IS NULL THEN 'UNKNOWN'
        WHEN SALARY >= 85000 THEN 'Senior'
        WHEN SALARY >= 65000 THEN 'Mid'
        WHEN SALARY >= 0 THEN 'Junior'
        ELSE 'UNKNOWN'
    END AS SALARY_BAND
FROM EMPLOYEES;

# 2
SELECT * FROM ORDERS;
SELECT ORDER_ID,
    AMOUNT,
    STATUS,
    CASE
        WHEN STATUS = 'Delivered' THEN 'Completed' 
        WHEN STATUS = 'Delivered' THEN 'Completed'
        WHEN STATUS = 'Shipped' THEN 'In Transit'
        WHEN STATUS = 'Pending' THEN 'Waiting'
        WHEN STATUS = 'Cancelled' THEN 'Closed'
        ELSE 'Unknown'
    END AS STATUS_FLAG
FROM ORDERS;

# SQL Exercise 2 - String Functions
'''
1. From employees:
   Display name in UPPER case.
   Display department in LOWER case.
   Show the length of each employee's name.
   Show first 5 characters of name using SUBSTR.

2. From customers:
   Concatenate first_name and last_name
   into a single column called full_name
   with a space between them.
   Use CONCAT or || operator.

3. From patients:
   Show names where city contains more
   than 5 characters.
   Use LENGTH().

4. From doctors:
   Remove the "Dr. " prefix from name column
   in the output. Do not update the table.
   Use REPLACE or SUBSTR.

Concepts: UPPER, LOWER, LENGTH, SUBSTR,
          CONCAT, ||, REPLACE, TRIM

---------------------------------------------
'''
# 1
SELECT
    UPPER(NAME) AS CAPITAL_LATTER,
    LOWER(DEPARTMENT) AS SMALL_LATTER,
    LENGTH(NAME) AS NAME_LENGTH,
    SUBSTR(NAME, 1, 5) AS NAME_FIRST5
FROM EMPLOYEES;

# 2
SELECT * FROM CUSTOMERS;
SELECT
    CONCAT(CONCAT(FIRST_NAME, ' '), LAST_NAME) AS FULL_NAME
FROM CUSTOMERS;

SELECT
    FIRST_NAME || ' ' || LAST_NAME AS FULL_NAME
FROM CUSTOMERS;

# 3
SELECT * FROM PATIENTS;
SELECT NAME, CITY FROM PATIENTS WHERE LENGTH(CITY) > 5;

# 4
SELECT
    REPLACE(name, 'Dr. ', '') AS name_clean
FROM doctors;

# SQL Exercise 3 - Combined multi-concept
'''
Write one query for each:

1. From employees:
   Show department, count of employees,
   average salary (rounded),
   and a dept_status column:
   avg salary >= 80000 → "High Pay"
   avg salary >= 60000 → "Mid Pay"
   below 60000         → "Low Pay"
   Use CASE inside SELECT after aggregation.
   Only include active employees.
   Order by average salary descending.

2. From appointments:
   Show doctor_id, count of appointments,
   total fee collected, average fee,
   and a workload_level column:
   count >= 3 → "High"
   count = 2  → "Medium"
   count = 1  → "Low"
   Use HAVING to exclude doctors with
   zero appointments.
   Order by total fee descending.

3. From orders and products (two tables):
   For each product, show:
   product_id, total orders, total quantity,
   total revenue, avg revenue per order.
   Only include products with
   total revenue above 500.
   Order by total revenue descending.
   Note: You will need a JOIN here.
   Use: orders o, products p
   WHERE o.product_id = p.product_id
   (implicit join — JOINs covered next week)
'''

# 1
SELECT * FROM EMPLOYEES;
SELECT DEPARTMENT,
    COUNT(*) AS EMP_COUNT,
    ROUND(AVG(SALARY)) AS AVG_SALARY,
    CASE
        WHEN (AVG(SALARY)) >= 80000 THEN 'High Pay'
        WHEN (AVG(SALARY)) >= 60000 THEN 'Mid Pay'
        ELSE 'Low Pay'
    END AS dept_status
FROM EMPLOYEES
WHERE IS_ACTIVE = 1
GROUP BY DEPARTMENT
ORDER BY AVG_SALARY DESC;

# 2
SELECT DOCTOR_ID,
    COUNT(*) AS APPT_COUNT,
    SUM(FEE) AS TOTAL_FEE,
    AVG(FEE) AS AVG_FEE,
    CASE
        WHEN COUNT(*) >= 3 THEN 'High'
        WHEN COUNT(*) = 2 THEN 'Medium'
        WHEN count(*) = 1 THEN 'Low'
        ELSE 'Unknow'
    END AS WORKLOAD_LEVEL 
From APPOINTMENTS
GROUP BY DOCTOR_ID
HAVING COUNT(*) > 0
ORDER BY TOTAL_FEE DESC;
```

```python
import pandas as pd

df_appointments = pd.read_csv("appointments.csv")
df_doctors = pd.read_csv("doctors.csv")
df_patients = pd.read_csv("patients.csv")
```

```python
# 1. Merge appointments with doctors on doctor_id.
#    Use pd.merge() with how='left'.
merged_df = pd.merge(
    df_appointments,
    df_doctors,
    on="doctor_id",
    how="left"
)
print(merged_df.columns)
```

```python

# 3. From the merged DataFrame:
#    - Find which specialty generated
#      the most total fee revenue.
#    - Find the average fee per specialty.
#    - Find which doctor had the most appointments.

total_fee_revenue = merged_df.groupby("specialty")["fee"].sum()

top_specialty = total_fee_revenue.idxmax()
print(f"Top Revenue Specialty: {top_specialty}")

ave_fee_per_specialty = merged_df.groupby("specialty")["fee"].mean()
print(ave_fee_per_specialty)

appt_count = merged_df.groupby("name")["appt_id"].count()

top_doctor = appt_count.idxmax()
print(f"Doctor with most appointments: {top_doctor}")
```

```python
# 4. Now merge appointments with patients
#    on patient_id.
#    Find the average fee paid per city.
merged_appt_and_pati = pd.merge(
    df_appointments,
    df_patients,
    on="patient_id",
    how="left"
)
print(merged_appt_and_pati.columns)
average_fee_per_city = merged_appt_and_pati.groupby("city")["fee_y"].mean()

print(average_fee_per_city)
```

```python
# Exercise 5 - Lambda and apply() revision
'''
Given this DataFrame:

import pandas as pd
data = {
    "name":   ["Raj", "Sneha", "Tom",
               "Fatima", "Liu", "Ana"],
    "age":    [34, 39, 46, 29, 24, 56],
    "fee":    [500, 200, 750, 600, 300, 900],
    "city":   ["Delhi","Mumbai","Chicago",
               "Dubai","Beijing","Madrid"]
}
df = pd.DataFrame(data)

1. Use apply() with lambda to create a column
   fee_in_inr converting USD fee to INR
   (multiply by 83).

2. Use apply() with lambda to create a column
   name_length with length of each name.

3. Use apply() with a named function to create
   a column called risk_category:
   age > 50 AND fee > 700 → "High Risk"
   age > 40              → "Medium Risk"
   otherwise             → "Low Risk"

4. Use apply() on the entire row (axis=1)
   to create a column called profile_tag
   combining name and city:
   Example: "Raj | Delhi"

Concepts: apply(), lambda, axis=0, axis=1,
          named function inside apply
'''
```

```python
import pandas as pd
data = {
    "name":   ["Raj", "Sneha", "Tom",
               "Fatima", "Liu", "Ana"],
    "age":    [34, 39, 46, 29, 24, 56],
    "fee":    [500, 200, 750, 600, 300, 900],
    "city":   ["Delhi","Mumbai","Chicago",
               "Dubai","Beijing","Madrid"]
}
df = pd.DataFrame(data)
print(df)
```

```python
# 1. Use apply() with lambda to create a columnS
#    fee_in_inr converting USD fee to INR
#    (multiply by 83).

df["fee_in_inr"] = df["fee"].apply(lambda x: x * 83)
print(df)
```

```python
# 2. Use apply() with lambda to create a column
#    name_length with length of each name.

df["name_length"] = df["name"].apply(lambda x: len(x))
print(df)
```

```python
# 3. Use apply() with a named function to create
#    a column called risk_category:
#    age > 50 AND fee > 700 → "High Risk"
#    age > 40              → "Medium Risk"
#    otherwise             → "Low Risk"

def risk_category_band(x):
    if x["age"] > 50 and x["fee"] > 700:
        return "High Risk"
    elif x["age"] > 40:
        return "Medium Risk"
    else:
        return "Low Risk"

df["risk_category"] = df.apply(risk_category_band, axis=1)
print(df)
```

```python
# 4. Use apply() on the entire row (axis=1)
#    to create a column called profile_tag
#    combining name and city:
#    Example: "Raj | Delhi"
df["profile_tag"] = df.apply(lambda x: f"{x["name"]} | {x["city"]}", axis=1)
print(df)

# Key Concepts
'''
apply() behavior
    axis=0 → column-wise (default)
    axis=1 → row-wise (used when multiple columns needed)
'''
# When to use what
'''
    Single column → df['col'].apply(lambda x: ...)
    Multiple columns → df.apply(func, axis=1)
'''
```

```python
# =================================================================
# Exercise 3 – Count Occurrence of a Character
'''
Ask the user to enter a word.

Ask the user to enter a character.

Count how many times that character appears in the word.

Example

Word: banana
Character: a

Output

3
⸻
'''
# Exercise 3 – Count Occurrence of a Character
words = input("Enter a word: ")
print(f"Word: {words}")
char = input("Enter a character: ")

count = 0
for i in words:
    if char == i:
        count += 1

print(f"count of \'{char}\' character in the word: {count}")

```

```python
# =================================================================
# Exercise 7 – Find Student With Highest Marks

'''
Given dictionary:

marks = {
Rahul:85,
Amit:72,
Priya:90,
Neha:78
}

Find the student who scored the highest marks.
⸻
'''
# Exercise 7 – Find Student With Highest Marks
marks = {
    "Rahul": 85,
    "Amit": 72,
    "Priya": 90,
    "Neha": 78
}
highest_mark = max(marks.values())
for key, value in marks.items():
    if value == highest_mark:
        print(f"{key} scored of highest marks: {highest_mark}")

```

```python
# =================================================================
# Exercise 9 – Read File and Print Line Numbers
'''
Create a file called notes.txt.

Add at least 5 lines of text.

Write a Python program to read the file and print line numbers with each line.

Example output

1 Python is powerful
2 Data engineering uses Python
3 Linux is important
⸻
'''

# Exercise 9 – Read File and Print Line Numbers
with open("notes.txt", "w") as file:
    file.write("Python is a popular programming language.\n")
    file.write("It is easy to learn and use.\n")
    file.write("Python is widely used in data analysis and automation.\n")
    file.write("It has many libraries like Pandas and NumPy.\n")
    file.write("Python is commonly used by data engineers and developers.")

with open("notes.txt", "r") as file:
    line_number = 1
    for lines in file:
        print(line_number, lines)
        line_number += 1
```

```python
# =================================================================
# Exercise 10 – Filter Data From File

'''
Create a file called scores.txt.

Example data:

Rahul 85
Amit 60
Priya 92
Neha 55
Karan 75

Write a Python program to display only students whose marks are greater than 70.

⸻
'''
# Exercise 10 – Filter Data From File
with open("scores.txt", "w") as file:
    file.write("Rahul 85\n")
    file.write("Amit 60\n")
    file.write("Priya 92\n")
    file.write("Neha 55\n")
    file.write("Karan 75\n")

with open("scores.txt", "r") as file:
    data = file.readlines()
    for i in data:
        data1 = i.split()
        name = data1[0]
        marks = int(data1[1])

        if marks > 70:
            print(name, marks)


```

```python
# =================================================================
# Exercise 11 – CSV File Reading
'''
Create a CSV file called employees.csv.

Example structure:

id,name,salary

1,Rahul,50000
2,Amit,45000
3,Priya,70000

Write a Python program to read and print each row.

Hint

Use csv module.

⸻
'''

# Exercise 11 – CSV File Reading

import csv

with open("employees.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Id", "Name", "Salary"])
    writer.writerow([1, "Rahul", 35000])
    writer.writerow([2, "Sneha", 42000])
    writer.writerow([3, "Amit", 60000])
    writer.writerow([4, "Priya", 38000])
    writer.writerow([5, "Rohit", 41000])
    writer.writerow([6, "Anjali", 65000])
    writer.writerow([7, "Karan", 36000])
    writer.writerow([8, "Meena", 43000])
    writer.writerow([9, "Vijay", 70000])
    writer.writerow([10, "Pooja", 39000])

with open("employees.csv", "r") as file:
    data = csv.reader(file)
    for row in data:
        print(row)
```

```python
# =================================================================
# Exercise 12 – Simple Data Cleaning
'''
Given list:

data = [“Rahul”,“Amit”,””,“Priya”,””,“Neha”]

Remove empty values and create a clean list.

Expected output

[“Rahul”,“Amit”,“Priya”,“Neha”]

⸻
'''
# Exercise 12 – Simple Data Cleaning
data = ["Rahul","Amit","","Priya","","Neha"]

clean_list = []
for i in data:
    if i != "":
        clean_list.append(i)
print(f"Clean list is: {clean_list}")
```

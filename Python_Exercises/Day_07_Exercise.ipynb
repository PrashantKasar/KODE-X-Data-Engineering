```python
# DAY 7 EXERCISES
# SQL PRACTICE
# Topic: Aggregation Revision and Practice
'''
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

Use this dataset. Create the table in SQLite
or work through the queries on paper.

Table: orders
order_id, customer_name, product, category,
quantity, unit_price, city, order_date

1   Rahul    Laptop    Electronics  2  55000  Pune       2024-01-05
2   Priya    Phone     Electronics  3  20000  Mumbai     2024-01-07
3   Amit     Desk      Furniture    1  12000  Pune       2024-01-10
4   Neha     Laptop    Electronics  1  55000  Delhi      2024-01-12
5   Karan    Chair     Furniture    4   4500  Bangalore  2024-01-15
6   Sneha    Phone     Electronics  2  20000  Pune       2024-01-18
7   Rohan    Bookshelf Furniture    2   8000  Mumbai     2024-01-20
8   Divya    Tablet    Electronics  3  15000  Delhi      2024-01-22
9   Meera    Desk      Furniture    1  12000  Pune       2024-01-25
10  Arjun    Laptop    Electronics  1  55000  Bangalore  2024-01-28

Add a computed column for total_amount
= quantity x unit_price when inserting,
or calculate it in your SELECT query.

---------------------------------------------
'''
CREATE TABLE ORDERS2(order_id NUMBER, customer_name VARCHAR2(20), product VARCHAR2(20), category VARCHAR2(20),
quantity NUMBER, unit_price NUMBER, city VARCHAR2(20), order_date DATE
);

INSERT INTO ORDERS2 VALUES (1,  'Rahul', 'Laptop', 'Electronics', 2, 55000, 'Pune', DATE '2024-01-05');
INSERT INTO ORDERS2 VALUES (2,  'Priya', 'Phone', 'Electronics', 3, 20000, 'Mumbai', DATE '2024-01-07');
INSERT INTO ORDERS2 VALUES (3,  'Amit', 'Desk', 'Furniture', 1, 12000, 'Pune', DATE '2024-01-10');
INSERT INTO ORDERS2 VALUES (4,  'Neha', 'Laptop', 'Electronics', 1, 55000, 'Delhi', DATE '2024-01-12');
INSERT INTO ORDERS2 VALUES (5,  'Karan', 'Chair', 'Furniture', 4, 4500, 'Bangalore', DATE '2024-01-15');
INSERT INTO ORDERS2 VALUES (6,  'Sneha', 'Phone', 'Electronics', 2, 20000, 'Pune', DATE '2024-01-18');
INSERT INTO ORDERS2 VALUES (7,  'Rohan', 'Bookshelf', 'Furniture', 2, 8000, 'Mumbai', DATE '2024-01-20');
INSERT INTO ORDERS2 VALUES (8,  'Divya', 'Tablet', 'Electronics', 3, 15000, 'Delhi', DATE '2024-01-22');
INSERT INTO ORDERS2 VALUES (9,  'Meera', 'Desk', 'Furniture', 1, 12000, 'Pune', DATE '2024-01-25');
INSERT INTO ORDERS2 VALUES (10, 'Arjun', 'Laptop', 'Electronics', 1, 55000, 'Bangalore', DATE '2024-01-28');

# SQL Exercise 1 - Aggregate Basics
# 
# 1. Count total number of orders.
# 2. Find total revenue across all orders.
#    total_amount = quantity x unit_price
# 3. Find average order value.
# 4. Find the most expensive unit price.
# 5. Find the cheapest unit price.
# 6. Count how many distinct cities placed orders.
# 7. Count how many distinct products were ordered.

-- 1. Count total number of orders.
SELECT COUNT(*) AS ORDERS_COUNT  FROM ORDERS2;

-- 2. Find total revenue across all orders.
--    total_amount = quantity x unit_price
SELECT SUM(QUANTITY * UNIT_PRICE) AS TOTAL_REVENUE FROM ORDERS2;

-- 3. Find average order value.
SELECT AVG(QUANTITY * UNIT_PRICE) AS AVERAGE_VALUE FROM ORDERS2;

-- 4. Find the most expensive unit price.
SELECT MAX(UNIT_PRICE) AS EXPENSIVE_UNIT_PRICE FROM ORDERS2;

-- 5. Find the cheapest unit price.
SELECT MIN(UNIT_PRICE) AS CHEAPEST_UNIT_PRICE FROM ORDERS2;

-- 6. Count how many distinct cities placed orders.
SELECT COUNT(DISTINCT CITY) AS COUNT FROM ORDERS2;

-- 7. Count how many distinct products were ordered.
SELECT COUNT(DISTINCT PRODUCT) AS COUNT_PRODUCT FROM ORDERS2;
SELECT * FROM ORDERS2;
---------------------------------------------

# SQL Exercise 2 - GROUP BY Revision

# 1. Total revenue per category.
# 2. Number of orders per city.
# 3. Average quantity ordered per product.
# 4. Total revenue per city, ordered highest first.
# 5. Number of orders per customer.
#    (Every customer here has one order,
#     but write the query correctly anyway.)

-- 1. Total revenue per category.
SELECT CATEGORY, SUM(QUANTITY * UNIT_PRICE) AS TOTAL_REVENUE FROM ORDERS2 GROUP BY CATEGORY;

-- 2. Number of orders per city.
SELECT CITY, COUNT(*) AS COUNT_ORDER_PER_CITY FROM ORDERS2 GROUP BY CITY;

-- 3. Average quantity ordered per product.
SELECT PRODUCT, AVG(QUANTITY * UNIT_PRICE) AS AVERAGE FROM ORDERS2 GROUP BY PRODUCT;

-- 4. Total revenue per city, ordered highest first.
SELECT CITY, SUM(QUANTITY * UNIT_PRICE) AS TOTAL_REVENUE FROM ORDERS2 GROUP BY CITY ORDER BY TOTAL_REVENUE DESC FETCH FIRST 1 ROW ONLY;

-- 5. Number of orders per customer.
--    (Every customer here has one order,
--     but write the query correctly anyway.)
SELECT CUSTOMER_NAME, COUNT(*) AS COUNT FROM ORDERS2 GROUP BY CUSTOMER_NAME;
---------------------------------------------

# SQL Exercise 3 - WHERE with GROUP BY

# 1. Total revenue per category,
#    only for orders from Pune.
# 2. Count of orders per city,
#    only for Electronics category.
# 3. Average unit price per category,
#    only for orders where quantity is more than 1.

-- 1. Total revenue per category,
--    only for orders from Pune.
SELECT CATEGORY, SUM(QUANTITY * UNIT_PRICE) AS TOTAL_REVENUE FROM ORDERS2 WHERE CITY = 'Pune' GROUP BY CATEGORY;
SELECT * FROM ORDERS2;

-- 2. Count of orders per city,
--    only for Electronics category.
SELECT CITY, COUNT(*) AS COUNT_OF_ORDER FROM ORDERS2 WHERE CATEGORY = 'Electronics' GROUP BY CITY;

-- 3. Average unit price per category,
--    only for orders where quantity is more than 1.
SELECT CATEGORY, AVG(UNIT_PRICE) AS AVERAGE FROM ORDERS2 WHERE QUANTITY > 1 GROUP BY CATEGORY;

---------------------------------------------

# SQL Exercise 4 - HAVING Revision

# 1. Show cities that have placed more than 1 order.
# 2. Show categories where total revenue
#    exceeds 100000.
# 3. Show products where average quantity ordered
#    is more than 1.
# 4. Show cities where total quantity ordered
#    is more than 3.
 
-- 1. Show cities that have placed more than 1 order.
SELECT CITY, COUNT(*) FROM ORDERS2 GROUP BY CITY HAVING COUNT(*) > 1;
SELECT * FROM ORDERS2;

-- 2. Show categories where total revenue
--    exceeds 100000.
SELECT CATEGORY, SUM(QUANTITY * UNIT_PRICE) AS TOTAL_REVENUE FROM ORDERS2 GROUP BY CATEGORY HAVING SUM(QUANTITY * UNIT_PRICE) > 100000;

-- 3. Show products where average quantity ordered
--    is more than 1.
SELECT PRODUCT, AVG(QUANTITY) AS AVERAGE FROM ORDERS2 GROUP BY PRODUCT HAVING AVG(QUANTITY) > 1;

-- 4. Show cities where total quantity ordered
--    is more than 3.
SELECT CITY, SUM(QUANTITY) AS TOTAL_QUANTITY FROM ORDERS2 GROUP BY CITY HAVING SUM(QUANTITY) > 3;
---------------------------------------------

# SQL Exercise 5 - Combined Challenge

# Write a single query that produces this summary:
# 
# - Category name
# - Total orders
# - Total quantity sold
# - Total revenue
# - Average order value (rounded to 0 decimals)

# Show only categories with total revenue above 50000.
# Order by total revenue descending.
# 
# Concepts: COUNT, SUM, AVG, ROUND,
#           GROUP BY, HAVING, ORDER BY

SELECT CATEGORY,
    COUNT(*) AS TOTAL_ORDER,
    SUM(QUANTITY) AS TOTAL_QUANTITY,
    SUM(QUANTITY * UNIT_PRICE) AS TOTAL_REVENUE,
    ROUND(AVG(QUANTITY * UNIT_PRICE), 0) AS AVERAGE
FROM ORDERS2
GROUP BY CATEGORY
HAVING SUM(QUANTITY * UNIT_PRICE) > 50000
ORDER BY AVERAGE DESC;

```

```python
# Exercise 3 - Functions Basics
'''
Write the following functions:

1. greet(name)
   Returns "Hello, Priya! Welcome to Data Engineering."
   Use f-string inside the function.

2. calculate_tax(salary)
   If salary > 50000, tax is 20 percent.
   If salary between 30000 and 50000, tax is 10 percent.
   Below 30000, no tax.
   Return the tax amount.

3. is_palindrome(word)
   Returns True if the word reads the same forwards
   and backwards. Ignore case.
   Test with: "Madam", "Python", "Racecar"

4. count_vowels(sentence)
   Returns the count of vowels in the sentence.

Call all four functions and print results
with clear labels.

Concepts: def, return, f-string, condition,
          slicing, lower(), loop
'''
# 1. greet(name)
#    Returns "Hello, Priya! Welcome to Data Engineering."
#    Use f-string inside the function.

def greet(name):
    print(f"Hello, {name} Welcome to Data Engineering.")

greet("Priya")

# 2. calculate_tax(salary)
#    If salary > 50000, tax is 20 percent.
#    If salary between 30000 and 50000, tax is 10 percent.
#    Below 30000, no tax.
#    Return the tax amount.

def calculate_tax(salary):
    if salary > 50000:
        print(f"tax: {salary * 0.02}")
    elif 30000 <= salary <= 50000:
        print(f"tax: {salary * 0.1}")
    else:
        print("No tax")

calculate_tax(25000)

# 3. is_palindrome(word)
#    Returns True if the word reads the same forwards
#    and backwards. Ignore case.
#    Test with: "Madam", "Python", "Racecar"

def is_palindrome(word):
    word = word.lower()
    if word == word[::-1]:
        return True
    else:
        return False

is_palindrome("Madam")

def is_palindrome1(word):
    word = word.lower()
    return word == word[::-1]

print(is_palindrome1("madam"))
print(is_palindrome1("PYTHON"))
print(is_palindrome1("Recear"))

# 4. count_vowels(sentence)
#   Returns the count of vowels in the sentence.
def count_vowels(sentence):
    count = 0
    vowels = ['a', 'e', 'i', 'o', 'u']
    for i in sentence:
        if i in vowels:
            count += 1
    print(f"Count of vowels is: {count}")

count_vowels("python is easy language.")
```

```python
# Exercise 5 - *args and **kwargs
'''
Part A - *args

Write a function called total_sales(*args) that:
- Accepts any number of sales figures
- Returns the total and average

Test with:
total_sales(1500, 2300, 4100)
total_sales(800, 1200, 3300, 2700, 4500)

Part B - **kwargs

Write a function called create_profile(**kwargs) that:
- Accepts any number of keyword arguments
- Prints each key and value on a new line

Test with:
create_profile(name="Rahul", city="Pune",
               role="Data Engineer", salary=72000)

Part C - Combined

Write a function called log_event(event_type,
*args, **kwargs) that:
- Prints the event type
- Prints all positional values as a list
- Prints all keyword values as key:value pairs

Test with:
log_event("pipeline_run", "step1", "step2",
          status="success", rows=5000)

Concepts: *args, **kwargs, loop, mixed arguments


---------------------------------------------
'''
# Part A - *args

# Write a function called total_sales(*args) that:
# - Accepts any number of sales figures
# - Returns the total and average

def total_sales(*args):
    print(f"Total Sum: {sum(args)}")
    print(f"Average: {sum(args) / len(args)}")

total_sales(1500, 2300, 4100)
total_sales(800, 1200, 3300, 2700, 4500)

# Part B - **kwargs

# Write a function called create_profile(**kwargs) that:
# - Accepts any number of keyword arguments
# - Prints each key and value on a new line

def create_profile(**kwargs):
    for key, value in kwargs.items():
        print(key, ':', value)

create_profile(name="Rahul", city="Pune", role="Data Engineer", salary=72000)

# Part C - Combined
# 
# Write a function called log_event(event_type,
# *args, **kwargs) that:
# - Prints the event type
# - Prints all positional values as a list
# - Prints all keyword values as key:value pairs
# 
# Test with:
# log_event("pipeline_run", "step1", "step2",
#           status="success", rows=5000)

def log_event(event_type, *args, **kwargs):
    print(f"Events type: {event_type}")
    print(f"Values:  {list(args)}")

    for key, value in kwargs.items():
        print(key, ':', value)

log_event("pipeline_run", "step1", "step2", status="success", rows=5000)
```

```python
# Exercise 6 - List Comprehensions
'''
Given this list of salaries:
salaries = [72000, 48000, 85000, 61000, 55000,
            52000, 47000, 91000, 38000, 67000]

Using list comprehensions (one line each):
1. Create a list of salaries above 60000.
2. Create a list of salaries after applying
   10 percent increment to each.
3. Create a list of "High" or "Low" for each salary
   based on whether it is above 60000 or not.
4. Create a list of squares of all salaries
   below 50000.

Then given this list of names:
names = ["rahul", "AMIT", "Priya", "NEHA", "karan"]

5. Create a new list with all names in Title Case.
6. Create a new list of names that have
   more than 4 characters.

Concepts: list comprehension, condition inside
          comprehension, string methods


---------------------------------------------
'''
salaries = [72000, 48000, 85000, 61000, 55000,
            52000, 47000, 91000, 38000, 67000]

# 1. Create a list of salaries above 60000.
above_60000_salary = [i for i in salaries if i > 60000]
print(f"Salary above 60000 list is: {above_60000_salary}")

# 2. Create a list of salaries after applying
#   10 percent increment to each.
increment = [round(i*1.10) for i in salaries]
print(f"After 10 percent increment list is: {increment}")

# 3. Create a list of "High" or "Low" for each salary
#    based on whether it is above 60000 or not.
salary = ["High" if i > 60000 else "Low" for i in salaries]
print(salary)

# 4. Create a list of squares of all salaries
#    below 50000.

squares = [i**2 for i in salaries if i < 50000]
print(f"Square: {squares}")

# Then given this list of names:
names = ["rahul", "AMIT", "Priya", "NEHA", "karan"]

# 5. Create a new list with all names in Title Case.
title_case = [i.title() for i in names]
print(f"Title case: {title_case}")

# 6. Create a new list of names that have
#   more than 4 characters.
only_4_characters_list = [char for char in names if len(char) > 4]
print(f"Only more than 4 characters list is: {only_4_characters_list}")
```

```python
# Exercise 7 - Dict Comprehensions
'''
Part A
Given two lists:
names   = ["Rahul", "Priya", "Amit", "Neha", "Karan"]
salaries = [72000, 85000, 48000, 61000, 55000]

Create a dictionary {name: salary} using
dict comprehension.

Part B
From the dictionary created above:
1. Create a new dict with only employees
   earning above 60000.
2. Create a new dict where each salary
   is increased by 5000.
3. Create a new dict where keys are names
   and values are "High" or "Low" based
   on salary threshold of 60000.

All in one line each using dict comprehension.

Concepts: dict comprehension, zip(),
          condition, arithmetic
'''
# Part A
# Given two lists:
# Create a dictionary {name: salary} using
# dict comprehension.
names   = ["Rahul", "Priya", "Amit", "Neha", "Karan"]
salaries = [72000, 85000, 48000, 61000, 55000]

dictionary = { Name : Salary for Name, Salary in zip(names,salaries)}
print(f"Dictionary: {dictionary}")

# Part B
# From the dictionary created above:
# 1. Create a new dict with only employees earning above 60000.
above_60000_salary = {Name: Salary for Name, Salary in zip(names, salaries) if Salary > 60000}
print(f"Only This Employees Earning Above 60000 dict is: {above_60000_salary}")

print("====="*9)
# 2. Create a new dict where each salary is increased by 5000.
New_Dict = {Name: Salary + 5000 for Name, Salary in zip(names,salaries)}
print(f"Increased 5000 salary dic is: {New_Dict}")

print("====="*9)
# 3. Create a new dict where keys are names
#    and values are "High" or "Low" based
#    on salary threshold of 60000.
New_Dict1 = {Name: "High" if Salary > 60000 else "Low" for Name, Salary in zip(names,salaries)}
print(New_Dict1)
```

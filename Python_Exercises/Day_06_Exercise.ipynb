```python
# SQL Exercise 2 - GROUP BY with WHERE
'''
Write queries for the following:

1. Count employees per department,
   but only include employees with salary above 50000.
   (Filter using WHERE before grouping.)

2. Find average salary per city,
   only for employees who joined after 2020.

3. Find total salary per department
   only for IT and Finance departments.
   Use WHERE with IN.

Note: WHERE filters rows before grouping.
      Remember this distinction clearly.

---------------------------------------------
'''
-- 1. Count employees per department,
--    but only include employees with salary above 50000.
--    (Filter using WHERE before grouping.)
SELECT DEPARTMENT, COUNT(DEPARTMENT) AS DEP_COUNT FROM EMPLOYEES06 
WHERE SALARY > 50000 
GROUP BY DEPARTMENT;

-- 2. Find average salary per city,
--    only for employees who joined after 2020.
SELECT CITY, AVG(SALARY) FROM EMPLOYEES06 
WHERE JOIN_YEAR > 2020 
GROUP BY CITY;

SELECT * FROM EMPLOYEES06;

-- 3. Find total salary per department
--    only for IT and Finance departments.
--    Use WHERE with IN.
SELECT DEPARTMENT, SUM(SALARY) FROM EMPLOYEES06
WHERE DEPARTMENT IN('IT','Finance') 
GROUP BY DEPARTMENT;
```

```python
# SQL Exercise 3 - HAVING
'''
Write queries for the following:

1. Show only departments that have more than
   2 employees.
2. Show departments where the average salary
   is above 60000.
3. Show cities where the total salary bill
   exceeds 150000.
4. Show departments where the minimum salary
   is below 50000.

Note: HAVING filters after grouping.
      WHERE filters before grouping.
      This difference is important.

---------------------------------------------
'''
-- 1. Show only departments that have more than
--    2 employees.

SELECT DEPARTMENT, COUNT(*) AS TOTAL_EMP 
FROM EMPLOYEES06 
GROUP BY DEPARTMENT
HAVING COUNT(*) > 2;

-- 2. Show departments where the average salary
--   is above 60000.
SELECT DEPARTMENT FROM EMPLOYEES06 
GROUP BY DEPARTMENT 
HAVING AVG(SALARY) > 60000;

-- 3. Show cities where the total salary bill
--   exceeds 150000.
SELECT CITY, SUM(SALARY) AS TOTAL_SALARY FROM EMPLOYEES06
GROUP BY CITY
HAVING SUM(SALARY) > 150000;

-- 4. Show departments where the minimum salary
--    is below 50000.
SELECT DEPARTMENT, MIN(SALARY) AS MINIMUM_SALARY FROM EMPLOYEES06
GROUP BY DEPARTMENT
HAVING MIN(SALARY) < 50000;
```

```python
# SQL Exercise 4 - Combined
'''
Write a single query that shows:
- Department name
- Number of employees
- Average salary (rounded to 0 decimal places)
- Highest salary
- Lowest salary

Only show departments with more than 1 employee.
Order results by average salary descending.

Concepts: COUNT, AVG, MAX, MIN, ROUND,
          GROUP BY, HAVING, ORDER BY
'''
-- # SQL Exercise 4 - Combined
-- Write a single query that shows:
-- - Department name
-- - Number of employees
-- - Average salary (rounded to 0 decimal places)
-- - Highest salary
-- - Lowest salary
-- 
-- Only show departments with more than 1 employee.
-- Order results by average salary descending.
SELECT DEPARTMENT, COUNT(*) AS TOTAL_COUNT, 
ROUND(AVG(SALARY), 0) AS AVG_SALARY, 
MAX(SALARY) AS MAXIMUM_SALARY, 
MIN(SALARY) AS MINIMUM_SALARY FROM EMPLOYEES06
GROUP BY DEPARTMENT
HAVING COUNT(*) > 1
ORDER BY AVG_SALARY DESC;

```

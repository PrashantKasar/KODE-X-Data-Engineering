```python
# ==============================================================================
# Exercise 2 – Function to Check Even or Odd
'''
Write a function called check_even_odd that takes a number as input.

If the number is even print “Even”.
If the number is odd print “Odd”.

Example:

check_even_odd(7)

Output

Odd
⸻
'''
# Exercise 2 – Function to Check Even or Odd
def check_even_odd(number):
    if number % 2 == 0:
        print("Even")
    else:
        print("Odd")

check_even_odd(7)

```

```python
# ==============================================================================
# Exercise 3 – Function to Find Largest Number
'''
Write a function that takes a list of numbers and returns the largest number.

Example list:

[10, 45, 2, 78, 33]

Expected output

Largest number: 78
⸻
'''
# Exercise 3 – Function to Find Largest Number
def largest_number(lst,n):
    # n = 0
    for i in lst:
        if i > n:
            n = i
    print(f"Largest Number: {n}")

largest_number([10, 45, 2, 78, 33], 0)

```

```python
# ==============================================================================
# Exercise 4 – Count Words in a Sentence
'''
Ask the user to enter a sentence.

Write a function that counts how many words are present in the sentence.

Example

Input
Python is easy to learn

Output
Number of words: 5
⸻

'''

# Exercise 4 – Count Words in a Sentence
def count_word(word):
    n = word.split()
    print(f"After split method input is: {n}")
    print(f"Number of words: {len(n)}")

word = input("Enter a sentence:")
print(f"Input: {word}")
count_word(word)
```

```python
# ==============================================================================
# Exercise 10 – Find Longest Word
'''
Ask the user to enter a sentence.

Write a function that finds the longest word in the sentence.

Example

Input
Python programming is powerful

Output
Longest word: programming
⸻
'''
# Exercise 10 – Find Longest Word
def longest_words(word):
    n = word.split()
    print(f"Longest word in the sentence is: \'{max(n, key=len)}\'")

word = input("Enter a sentence: ")
print(f"Input: {word}")
longest_words(word)

              
```

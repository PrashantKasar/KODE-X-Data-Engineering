```python
# ================================================================================
# PYTHON CHALLENGE
'''
Exercise 1 - Sum of Numbers

Ask user to enter 5 numbers.
Store them in a list.
Print:
Total sum
Average

⸻
'''
# Exercise 1 - Sum of Numbers
num = [int(input()) for i in range(5)]

print(f" Total sum is : {sum(num)}")
print(f"Average : {sum(num) / len(num)}")


```

```python
# ================================================================================
'''
Exercise 2 – Largest Number
Given list:
Find the largest number using a loop.
'''
list_1 = [5,12,7,20,3]
max_num = list_1[0]

for i in list_1:
    if i > max_num:
        max_num = i
print(f"Largest Number: {max_num}")
```

```python
# ================================================================================
'''
Exercise 3 – Count Even Numbers

Given list:
[2,5,8,11,14,17]
Count how many even numbers exist.
⸻
'''
# Exercise 3 – Count Even Numbers

list = [2,5,8,11,14,17]

count = 0
for i in list:
    if i % 2 == 0:
        count += 1
print(f"Count Even Number is : {count}")

```

```python
# ================================================================================
'''
Exercise 4 – String Reverse

Ask user to enter a word.
Print the reverse of the word.

Example:

Input: python
Output: nohtyp
⸻
'''

# Exercise 4 – String Reverse

word = input("Enter your word :")
r = word[::-1]
print(r)
```

```python
'''
Exercise 5 – Word Count

Ask user to enter a sentence.
Count number of words in the sentence.

⸻
'''
# ================================================================================
# Exercise 5 – Word Count

sentence = input("Enter your sentence :")
print(f"Input : {sentence}")

lst = sentence.split(',')
print(f"After convert the list : {lst}")

count = len(lst)
print(f"Count number of words is : {count}")

```

```python
'''
Exercise 6 – Duplicate Finder

Given list:
[1,2,3,2,4,5,3]
Print duplicate numbers.

Expected output:
2
3
# ================================================================================
⸻
'''
# Exercise 6 – Duplicate Finder

list = [1,2,3,2,4,5,3]
dup_item = set()
for i in list:
    if i in dup_item:
        print(i)
    else:
        dup_item.add(i)
```

```python
'''
Exercise 7 – Dictionary Iteration

Dictionary:

marks =
Rahul : 85
Amit : 78
Priya : 92

Print:
Rahul scored 85
Amit scored 78
Priya scored 92
⸻
# ================================================================================
'''
# Exercise 7 – Dictionary Iteration

marks = {
    "Rahul":85,
    "Amit": 78,
    "Priya":92
}
for key, value in marks.items():
    print(f"{key} scored {value}.")

```

```python
# ================================================================================
# DAY 02 - SQL CHALLENGE 1
# DATE - 06-03-2026

'''
Table: students, create and add some sample records

Columns:
id
name
marks

Write SQL queries to:
	1.	Show all students
	2.	Show students with marks greater than 70
	3.	Sort students by marks in descending order
	4.	Show top 3 students

⸻
-- Table: students, create and add some sample records
-- Columns: id, name, marks
'''
CREATE TABLE STUDENTS(id NUMBER, name VARCHAR2(20), marks NUMBER);
INSERT INTO STUDENTS VALUES(1, 'PRASHANT', 75);
INSERT INTO STUDENTS VALUES(2, 'VIJAY', 65);
INSERT INTO STUDENTS VALUES(3, 'AMIT', 89);
INSERT INTO STUDENTS VALUES(4, 'RAHUL', 95);


-- 1.	Show all students
SELECT * FROM STUDENTS;

--2 Show students with marks greater than 70
SELECT NAME, MARKS FROM STUDENTS WHERE MARKS > 70;

-- 3.	Sort students by marks in descending order
SELECT * FROM STUDENTS ORDER BY MARKS DESC;

--	4.	Show top 3 students
SELECT * FROM STUDENTS ORDER BY MARKS DESC FETCH FIRST 3 ROW ONLY;

# ================================================================================

# SQL CHALLENGE 2

'''
Table: orders, create and sample records

Columns:
order_id
customer
amount

Write SQL queries to:
	1.	Find total number of orders
	2.	Find average order amount
	3.	Find customer with highest order amount
	4.	Find orders greater than 500

Use:

COUNT
AVG
ORDER BY
WHERE

⸻
-- Table: orders, create and sample records
-- Columns: order_id, customer, amount
'''

CREATE TABLE ORDERS1(order_id NUMBER, customer_name VARCHAR2(20), amount NUMBER);
INSERT INTO ORDERS1 VALUES(101, 'RAHUL', 45000);
INSERT INTO ORDERS1 VALUES(102, 'PRATIK', 45000);
INSERT INTO ORDERS1 VALUES(103, 'VIJAY', 85000);
INSERT INTO ORDERS1 VALUES(104, 'VISHAL', 4000);
INSERT INTO ORDERS1 VALUES(105, 'RAM', 35000);
SELECT * FROM ORDERS1;

-- 1.	Find total number of orders
SELECT COUNT(*) FROM ORDERS1;

-- 	2.	Find average order amount
SELECT AVG(AMOUNT) FROM ORDERS1;

-- 3.	Find customer with highest order amount
SELECT * FROM ORDERS1 ORDER BY AMOUNT DESC FETCH FIRST 1 ROW ONLY;

-- 4.	Find orders greater than 500
SELECT ORDER_ID, AMOUNT FROM ORDERS1 WHERE AMOUNT > 500;

```

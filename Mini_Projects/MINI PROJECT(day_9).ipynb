```python
# Task B - Clean doctors.csv:
# 1. Load raw/doctors.csv.
import pandas as pd

df = pd.read_csv("doctors.csv")
df.head()

# 2. Strip whitespace from all string columns.
df = df.apply(lambda col: col.str.strip() if col.dtype == object else col)
print(df)

# 3. Add a column seniority:
#    years >= 15 → "Principal"
#    years >= 10 → "Senior"
#    below 10    → "Associate"
import numpy as np

if "experience_years" in df.columns:
    conditions = [
        df["experience_years"] >= 15,
        df["experience_years"] >= 10,
        df["experience_years"] < 10
    ]

    choices = ["Principal", "Senior", "Associate"]

    df["Seniority"] = np.select(conditions, choices, default="Unknown")
    print(df)

# 4. Save to processed/doctors_clean.csv
#    without index.
df.to_csv("processed/doctors_clean.csv", index=False)

df.head()

import pandas as pd
import numpy as np

def clean_doctors():
    df = pd.read_csv("doctors.csv")
    print(df.head())

    df = df.apply(lambda col: col.str.strip() if col.dtype == object else col)
    print("-----------------------------------------------------")
    if "experience_years" in df.columns:
        conditions = [
            df["experience_years"] >= 15,
            df["experience_years"] >= 10,
            df["experience_years"] < 10
        ]

        choices = ["Principal", "Senior", "Associate"]
    
        df["Seniority"] = np.select(conditions, choices, default="Unknown")
        print(df.head())

        df.to_csv("processed/doctors_clean.csv", index=False)

        print("\nCleaned file saved to processed/doctors_clean.csv")


if __name__ == "__main__":
    clean_doctors()

```

```python
# Task C - Clean appointments.csv:

# 1. Load raw/appointments.csv.
import pandas as pd
import numpy as np

df = pd.read_csv("appointments.csv")
df.head(11)

# 2. Strip whitespace from all string columns.
df = df.apply(lambda col: col.str.strip() if col.dtype == object else col)
print(df)

# 3. Convert appt_date to datetime using
#    pd.to_datetime().
if "appt_date" in df.columns:
    df["appt_date"] = pd.to_datetime(df["appt_date"], errors="coerce")
    print(df)

# 4. Add a column fee_category:
#    fee >= 700 → "High"
#    fee >= 400 → "Medium"
#    fee < 400  → "Low"
if "fee" in df.columns:
    conditions = [
        df["fee"] >= 700,
        df["fee"] >= 400,
        df["fee"] < 400
    ]

    choices = ["High", "Medium", "Low"]
    df["fee_category"] = np.select(conditions, choices, default="Unknown")
    print(df)

# 5. Add a column day_of_week extracted from
#    appt_date.
if "appt_date" in df.columns:
    df["day_of_week"] = df["appt_date"].dt.day_name()
    print(df)
#    Hint: df['appt_date'].dt.day_name()


# 6. Save to processed/appointments_clean.csv
#    without index
df.to_csv("processed/appointments_clean.csv")

df.head(11)

```

```python
# STEP 3 - Python: Enrich by merging

# Create: processed/enricher.py
 
# 1. Load all three processed clean CSV files.
import pandas as pd

patients = pd.read_csv("processed/patients_clean.csv")
doctors = pd.read_csv("processed/doctors_clean.csv")
appointments = pd.read_csv("processed/appointments_clean.csv")


# 2. Merge appointments with patients
#    on patient_id (left join).
df = pd.merge(
    appointments,
    patients,
    on="patient_id",
    how="left"
)
# print(df.columns)

# 3. Merge the result with doctors
#    on doctor_id (left join).
df= pd.merge(
    df,
    doctors,
    on="doctor_id",
    how="left"
)
#print(df.columns)


# 4. The final merged DataFrame should have:
#    appt_id, patient name, patient age_group,
#    patient city, doctor name, specialty,
#    seniority, appt_date, day_of_week,
#    diagnosis, fee, fee_category.
# 5. Select only these meaningful columns
#    and rename them cleanly if needed.
df_final = df[[
    'appt_id',
    'appt_date',
    'patient_id',
    'name_x',
    'age',
    'city',
    'doctor_id',
    'name_y',
    'specialty',
    'experience_years',
    'diagnosis',
    'fee',
    'fee_category',
    'day_of_week'
]].rename(columns={
    'name_x': 'patient_name',
    'name_y': 'doctor_name'
})
print(df_final.columns)

# 6. Save to processed/enriched_data.csv
#    without index.
df_final.to_csv("processed/enriched_data.csv")

# 7. Print shape and first 5 rows to verify.
print("Shape:", df_final.shape)

df.head(10)
```

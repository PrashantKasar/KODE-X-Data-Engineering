```python

STEP 2 - Python Task Answer -

#Function 1: read_orders(filename)

def read_orders(filename):
    with open(filename, "r", newline="") as file:
        data = file.readlines()
        header = data[0].strip().split(',')

        order = [dict(zip(header,line.strip().split(','))) for line in data[1:]]
    return order

# Function 2: total_revenue(orders)

def total_revenue(orders):
    return sum(int(data['quantity']) * int(data['unit_price']) for data in orders)

# Function 3: revenue_by_category(orders)

def revenue_by_category(orders):
    dict_category = {}
    for i in orders:
        if i['category'] not in dict_category:
            dict_category[i['category']] = 0
        
        dict_category[i['category']] += int(i['quantity']) * int(i['unit_price'])
    return dict_category
            
# Function 4: top_customers(orders, n=3)

def top_customers(orders,n=3):
    top_customers = {}
    for i in orders:
        if i['customer_name'] not in top_customers:
            top_customers[i['customer_name']] = 0

        top_customers[i['customer_name']] = int(i['quantity']) * int(i['unit_price'])
        sorted_cus = sorted(top_customers.items(), key=lambda x: [x[1]],reverse=True)
    return sorted_cus[:n]

# Function 5: city_order_count(orders)

def city_order_count(orders):
    city_count = {}
    for i in orders:
        if i['city'] not in city_count:
            city_count[i['city']] = 0

        city_count[i['city']] += 1
    return city_count
# Function 6: print_report(orders)

def  print_report(orders, *args, **kwargs):
    print("\n SUMMARY REPORT")
    print("====="*5)

    print(f"Total Revenue: {total_revenue(orders)}")
    print("\nRevenue By Category:")
    for key, value in revenue_by_category(orders).items():
        print(key, ':', value)

    n = kwargs.get("top_n", 3)
    print(f"\nTop {n} Customers:")
    for key,value in top_customers(orders,n):
        print(key, ':', value)

    print("\nOrders By City:")
    for key,value in city_order_count(orders).items():
        print(key, ':', value)

orders = read_orders("orders.csv")
print_report(orders,top_n=3)
```

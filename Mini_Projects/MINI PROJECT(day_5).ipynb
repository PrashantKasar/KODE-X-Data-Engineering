```python
# Day 05
# STEP 3 – Python Task

# 1.Read the students.csv file.
import csv

with open("students.csv", "r") as file:
    data = csv.reader(file)
    for row in data:
        print(row)
# 2.Remove students whose marks are less than 70.
clean_data = []
with open("students.csv", "r") as file:
    read_data = csv.DictReader(file)
    for row in read_data:
        if int(row['marks']) > 70:
            number = row.copy()
            number['name'] = number['name'].upper()
            clean_data.append(number)
# 3.Convert student names to uppercase.
# 4.Save the cleaned data into: clean_students.csv
with open("clean_students.csv", "w") as file:
    writer = csv.DictWriter(file, fieldnames= ['id', 'name', 'marks', 'city'])
    writer.writeheader()
    writer.writerows(clean_data)
    
with open("clean_students.csv", "r") as file:
    read_data = csv.DictReader(file)
    for row in read_data:
        print(row)
```

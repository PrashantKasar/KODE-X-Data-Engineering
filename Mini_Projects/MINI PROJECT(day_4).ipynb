```python
# STEP 1 – Create Project Folder (Linux)
mkdir -p student_project
cd student_project/
touch students.csv clean_students.csv analysis.sql
```

```python
# STEP 2 – Create Sample Dataset
cat > students.csv << EOF
id,name,age,marks,city
1,Rahul,21,85,Pune
2,Amit,22,72,Mumbai
3,Priya,21,90,Pune
4,Neha,23,65,Delhi
5,Karan,22,78,Bangalore
EOF

 cat students.csv
 cat students.csv | column -s, -t
```

```python
# STEP 3 – Python Task

# Read the original file
with open(r'D:\Kode-x notes\linux\student_project\students.csv', "r", newline="") as file:
    lines = file.readlines()

# Store cleaned data
cleaned_data = []

# Keep the header
cleaned_data.append(lines[0])

# Process each student's record
for line in lines[1:]:
    student = line.strip().split(",")

    student_id = student[0]
    name = student[1].upper()
    age = student[2]      # Convert name to uppercase
    marks = int(student[3])
    city = student[4]

    # Keep only students with marks >= 70
    if marks >= 70:
        cleaned_data.append(f"{student_id},{name},{age},{marks},{city}\n")

# Save the cleaned data to a new file
with open("cleaned_students.csv", "w") as file:
    file.writelines(cleaned_data)

print("Cleaned data saved to cleaned_students.csv")
```

```python
# STEP 4 – SQL Task

CREATE TABLE STU_DATA(id NUMBER, name VARCHAR2(20), age NUMBER, marks FLOAT, city VARCHAR2(20));
INSERT INTO STU_DATA VALUES(1,'Rahul',21,85,'Pune');
INSERT INTO STU_DATA VALUES(2,'Amit',22,72,'Mumbai');
INSERT INTO STU_DATA VALUES(3,'Priya',21,90,'Pune');
INSERT INTO STU_DATA VALUES(4,'Neha',23,65,'Delhi');
INSERT INTO STU_DATA VALUES(5,'Karan',22,78,'Bangalore');

SELECT * FROM STU_DATA;

SELECT AVG(MARKS) AS AVERAGE_MARKS FROM STU_DATA;

SELECT MAX(MARKS) AS HIGHEST_MARKS FROM STU_DATA

SELECT * FROM STU_DATA ORDER BY MARKS DESC FETCH FIRST 1 ROW ONLY;

SELECT
    CITY,
    COUNT(*) AS NO_OF_STUDENT
FROM STU_DATA
GROUP BY CITY;
```

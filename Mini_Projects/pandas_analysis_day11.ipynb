```python
# Task 2 — Clean and enrich employees:
# - Strip whitespace from string columns.
# - Add salary_band using apply():
#   >= 85000 → "Senior"
#   >= 65000 → "Mid"
#   below    → "Junior"
# - Add experience_band using apply():
#   years >= 7 → "Veteran"
#   years >= 4 → "Experienced"
#   below 4    → "Fresher"
```

```python
result = df["name"] = df["name"].str.strip()
result1 = df["dept"] = df["dept"].str.strip()
```

```python
result = df.apply(lambda column: column.str.strip() if column.dtype == "object" else column)

print(result)
```

```python
def sal_band(s):
    if s >= 85000:
        return "Senior"
    elif s > 65000:
        return "Mid"
    else:
        return "Junior"
    
df["salary_band"] = df["salary"].apply(sal_band)
```

```python
print(df)
```

```python
def band(y):
    if y >= 7:
        return "Veteran"
    elif y >= 4:
        return "Experienced"
    else:
        return "Fresher"
    
df["experience_band"] = df["years"].apply(band)
```

```python
print(df)
```

```python
# Task 3 — Merge with departments:
# - Left join employees with departments
#   on dept column.
# - Keep: name, dept, salary, years, city,
#         budget, location, salary_band,
#         experience_band.
# - Save to processed/enriched_employees.csv
```

```python
left_join_df = pd.merge(df, df1, on="dept", how="left")
```

```python
print(left_join_df)
```

```python
left_join_df.to_csv("../processed/enriched_employees.csv")
```

```python
'''
Task 4 — Analysis (save each result):
- Dept summary: headcount, avg_salary,
  total_salary, budget, variance
  (budget minus total_salary).
  Save to reports/dept_summary.csv

- Salary band distribution:
  count per band.
  Save to reports/band_distribution.csv

- Top 3 earners per department.
  Save to reports/top_earners.csv

- City-wise headcount and avg salary.
  Save to reports/city_summary.csv
'''
```

```python
dept_summary = left_join_df.groupby('dept').agg( 
    headcount = ('emp_id', 'count'),
    avg_salary = ('salary', 'mean'),
    total_salary = ('salary', 'sum'),
    total_budget = ('budget', 'sum'),
    budget = ('budget', 'first')
).reset_index()
```

```python
dept_summary["variance"] = dept_summary["budget"] - dept_summary["total_salary"]
```

```python
dept_summary.to_csv("../reports/dept_summary.csv")
```

```python
band_distribution = df.groupby("salary_band").agg(
    emp_count = ('emp_id', 'count')
).reset_index()
```

```python
band_distribution.to_csv("../reports/band_distribution.csv", index=False)
```

```python
top_3_earners = (
    df.sort_values(
        ["dept", "salary"],
        ascending= [True, False]
    ).groupby("dept").head(3)
)
```

```python
top_3_earners.to_csv("../reports/top_earners.csv", index=False)
```

```python
city_summary = df.groupby("city").agg(
    headcount = ('emp_id', 'count'),
    avg_salary = ('salary', 'mean')
).reset_index()
```

```python
city_summary.to_csv("../reports/city_summary.csv", index=False)
```

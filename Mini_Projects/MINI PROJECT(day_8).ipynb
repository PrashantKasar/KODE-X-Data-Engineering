```python
def code_divider():
    print("-----------------------------------------------------------------------")

# 1. Reads all three CSV files using pandas.

import pandas as pd

def profile(df, name):
    code_divider()
    print(f"Filename: {name}")

    print(f"Shape: {df.shape[0]} rows x {df.shape[1]} columns")

    print("\nColumns:")
    print(list(df.columns))

    print("\nData Types:")
    print(df.dtypes)

    print("\nNull Values:")
    print(df.isnull().sum())

    print("\nFirst 3 Rows:")
    print(df.head(3))

def main():
    patients = pd.read_csv("patients.csv")
    doctors = pd.read_csv("doctors.csv")
    appointment = pd.read_csv("appointments.csv")

    profile(patients, "Patients")
    profile(doctors, "Doctors")
    profile(appointment, "Appointments")

    code_divider()
    print("Appointments Analysis")

    total_fee = appointment["fee"].sum()
    avg_fee = appointment["fee"].mean()
    min_fee = appointment["fee"].min()
    max_fee = appointment["fee"].max()

    print(f"Total Fee Collected: {total_fee}")
    print(f"Average Fee: {avg_fee}")
    print(f"Minmum Fee: {min_fee}")
    print(f"Maximum Fee: {max_fee}")

    doctor_counts = appointment.groupby("doctor_id").size()

    print("\nAppointments per Doctor:")
    print(doctor_counts)

    top_doctor = doctor_counts.idxmax()
    top_visits = doctor_counts.max()

    code_divider()
    print("Hospital Pipeline - Day 8 Profiling Complete")

    print(f"Patients : {len(patients)} records")
    print(f"Doctors : {len(doctors)} records")
    print(f"Appointments: {len(appointment)} records")
    print(f"Total Revenue: {total_fee}")
    print(f"Avg Fee     : {round(avg_fee, 2)}")
    print(f"Top Doctor by Appointments: doctor_id {top_doctor} ({top_visits} visits)")

if __name__ == "__main__":
    main()
```
